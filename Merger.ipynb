```python
import pandas as pd
from datetime import datetime as dt
from datetime import timedelta
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
btc = pd.read_csv('API Data/BTC_price.csv',index_col = 0)
btc = btc[~btc.index.duplicated()].reset_index()
doge = doge = pd.read_csv('API Data/DOGE_price.csv',index_col = 0)
doge = doge[~doge.index.duplicated()].reset_index()
tweet_doge = pd.read_csv('DOGE Sentiment.csv',index_col = 0)
tweet_btc = pd.read_csv('BTC Sentiment.csv',index_col = 0)
```

```python
tweet_doge
```

```python
doge
```

```python
all_times = pd.date_range(start = doge.time.min(), end = doge.time.max(), freq = '1H')
all_times = pd.DataFrame(all_times)
all_times['time'] = ''
for i in range(len(all_times[0])):
    all_times['time'][i] = str(all_times[0][i])
all_times.drop(0, axis = 1,inplace = True)
```

```python
doge = pd.merge(all_times,doge, on = "time", how = 'left')
doge.fillna(method="bfill",inplace = True)
doge = pd.merge(doge, tweet_doge, on="time", how = 'left')
doge.fillna(0,inplace = True)
```

```python
doge
```

```python
btc = pd.merge(all_times,btc, on = "time", how = 'left')
btc.fillna(method="bfill",inplace = True)
btc = pd.merge(btc, tweet_btc, on="time", how = 'left')
btc.fillna(0,inplace = True)
```

```python
btc
```

```python
features = ['close','volume','sentiment']
```

```python
btc[features].to_csv('BTC Cleaned.csv')
doge[features].to_csv('DOGE Cleaned.csv')
```
